```python
from dependencies import *
import os
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import nltk
```

### Load data and create index

```python
data = load_tweets()
tweets_cleaned = [clean_text(text) for text in data["text"]]
data['tweets_cleaned'] = tweets_cleaned

index, tf, df, idf, titleIndex = create_index(data['join'])
```

### Word2vec

```python
model = Word2Vec(window=5, min_count=1, workers=4,size=10)
model.build_vocab(tweets_cleaned)
model.train(tweets_cleaned,total_examples=len(tweets_cleaned),epochs=10)
```

```python
def tweet2vec(text,word2vec):
    if(text):
        vectors = word2vec[text[0]]
        for word in text[1:]:
            word_vector = word2vec[word]
            vectors = np.concatenate([vectors,word_vector])
        return vectors.reshape(len(text),-1).mean(axis=0)
    else:
        return


def query2vec(query, word2vec):

    vectors = -1
    for i,word in enumerate(query):
        #print(word)
        
        word_vector = word2vec[word]
        #print(word_vector.shape)
        if i  > 0:
            vectors = np.concatenate([vectors,word_vector])
        if i == 0:
            vectors = word_vector  
    return vectors.reshape(len(query),-1).mean(axis=0)
```

```python
representation = [tweet2vec(items,model) for items in tweets_cleaned]
embeded_tweets = np.array(representation)
data['embeded_tweets'] = list(embeded_tweets)

aux = []
for i,d in zip(data.index,data['embeded_tweets']):
    if d is None:
        aux.append(i)
data.drop(aux, inplace=True)
data.reset_index(drop = True, inplace=True)
```

```python
embeded_tweets = []
for i in data['embeded_tweets'].values:
    aux = []
    for j in i:
        aux.append(j)
    embeded_tweets.append(aux)
```

### TSNE

```python
tsne_model_en_2d = TSNE(perplexity=15, n_components=2, init='pca', n_iter=250, random_state=32)
embeddings_en_2d = np.array(tsne_model_en_2d.fit_transform(embeded_tweets))
```

```python
x = embeddings_en_2d[:, 0]
y = embeddings_en_2d[:, 1]
plt.figure(figsize=(20,10))
plt.scatter(x, y, c='red')
```

```python
CLUSTERS = 3
kmeans = KMeans(n_clusters=CLUSTERS, random_state=0,max_iter=1000).fit(embeddings_en_2d)
data["cluster"] = kmeans.labels_

grouped = data.groupby(by = data['cluster'])
for cluster in grouped.groups.keys():
    df = grouped.get_group(cluster)
    tweets_cleaned = [clean_text(text) for text in df["text"]]
    text = [text for text in tweets_cleaned]
    text = [text for text in text for text in text]
    freq = nltk.FreqDist(text)
    print(freq.most_common(5))
    print("\n")
```

```python
emb= pd.DataFrame([x,y,kmeans.labels_]).T
emb.rename(columns={0:'x', 1:'y', 2:'cluster'}, inplace=True)
sns.scatterplot(data=emb, x='x', y='y', hue='cluster')
```

## Diversification

```python
def diversity_score(tweets,num_clusters=3):
    top = len(tweets)
    best = num_clusters*np.sqrt(top/num_clusters)/top
    score = 0
    try:
        for c in tweets.unique():
            size = len(tweets[tweets==c])
            score += np.sqrt(size)
        #return abs(abs(score-best)%best-best)/best
        return (score/best)%1
    except:
        tweets = tweets['cluster']
        for c in tweets.unique():
            size = len(tweets[tweets==c])
            score += np.sqrt(size)
        #return abs(abs(score-best)%best-best)/best
        return (score/best)%1
```

```python
def diversify(data):
    top = len(data)
    clusters = len(data['cluster'].unique())
    result = pd.DataFrame()
    
    k = int(top/clusters)
    
    for cluster in range(clusters):
        aux = data[data['cluster']==cluster]
        result = pd.concat([result, aux.iloc[:k]])
    return result
```

```python
def google2(query, index, tf, df, idf, titleIndex, data, top=50):
    print('Query: {}'.format(query))

    ranked_docs = search_tf_idf(query, index, tf, df, idf, titleIndex, data)    
    results = pd.DataFrame(columns=['tweet', 'username', 'date', 'likes', 'retweets', 'url'])

    count = 0
    for d_id in ranked_docs[:top] :
        results.loc[count,'tweet'] = data.loc[d_id, "text"]
        results.loc[count,'username'] = data.loc[d_id, "user"]['screen_name']
        results.loc[count,'date'] = data.loc[d_id, "created_at"]
        results.loc[count,'likes'] = data.loc[d_id, "favorite_count"]
        results.loc[count,'retweets'] = data.loc[d_id, "retweet_count"]
        results.loc[count,'url'] = "https://twitter.com/twitter/statuses/"+str(data.loc[d_id, "id"])
        try:
            results.loc[count,'cluster'] = data.loc[d_id, "cluster"]
        except:
            print('An exception ocurred')
        count +=1
    return results
```

## RQ 2A - RQ 2B

```python
queries = ['pfizer', 'vaccine', 'biden', 'lockdowns']
```

```python
standard= []
diverse = []

for query in queries:
    results1 = google2(query, index, tf, df, idf, titleIndex, data)
    standard.append(diversity_score(results1['cluster']))
    
    results2 = diversify(results1)
    diverse.append(diversity_score(results2['cluster']))
    
plt.plot(standard, label='Without diversity')
plt.plot(diverse, label='With diversity')
plt.legend()
plt.show()
```
